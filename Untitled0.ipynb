```python
def binary_mask(pixel, low_threshold=22, treshold=41):
  if pixel >= low_threshold:
    return 0
  else:
    return 255
```

```python
# definitiva
exposure_image = Image.fromarray(exposure_image_array)
binary_image = exposure_image.point(lambda x: binary_mask(x, low_threshold=25))
binary_image_array = np.asarray(binary_image)
```

```python
# create figure
fig = plt.figure(figsize=(10, 7))
# setting values to rows and column variables
rows = 1
columns = 2
# Adds a subplot
fig.add_subplot(rows, columns, 1)
plt.imshow(array1, cmap='gray')
plt.axis('off')
plt.title("ORIGINAL TANQUE 23")
# Adds a subplot
fig.add_subplot(rows, columns, 2)
plt.imshow(binary_image_array, cmap='gray')
plt.axis('off')
plt.title("Aplicando Máscara Binaria")
```

```python
# Conteo de objetos en la imagen
img=measure.label(binary_image_array, background=0)
propsa = measure.regionprops(img)
length = len(propsa)
print ('Objetos contados = '+str(length))
```

```python
#DETECCION DE BORDES
from skimage.feature import canny
imcanny = canny(binary_image_array)
# Adds a subplot
plt.imshow(imcanny, cmap='gray')
plt.axis('off')
plt.title("Levaduras")
```

```python
#PREWITT

from skimage.filters import prewitt
imprewitt = prewitt(binary_image_array)
# Adds a subplot
plt.imshow(imprewitt, cmap='gray')
plt.axis('off')
plt.title("Prewitt")
```

```python
#FILTROS MORFOLOGICOS

fill_coins = ndi.binary_fill_holes(imprewitt)
rem_small = morphology.remove_small_objects(fill_coins, 40)
ero = morphology.erosion(rem_small, disk(4))
dil = morphology.dilation(ero, disk(2))
```

```python
# Aplicamos cierre sobre la imagen original
im3_original = closing(imprewitt, disk(5))
# create figure
fig = plt.figure(figsize=(14, 7))
# setting values to rows and column variables
rows = 2
columns = 3
# Adds a subplot
fig.add_subplot(rows, columns, 1)
plt.imshow(img1, cmap='gray')
plt.title("Original")
plt.axis('off')
# Adds a subplot
fig.add_subplot(rows, columns, 2)
plt.imshow(imprewitt, cmap='gray')
plt.title("Prewitt")
plt.axis('off')

# Adds a subplot
fig.add_subplot(rows, columns, 3)
plt.imshow(fill_coins, cmap='gray')
plt.title("Rellenado")
plt.axis('off')
# Adds a subplot
fig.add_subplot(rows, columns, 4)
plt.imshow(rem_small, cmap='gray')
plt.title("No estructuras pequeñas")
plt.axis('off')
# Adds a subplot
fig.add_subplot(rows, columns, 5)
plt.imshow(ero, cmap='gray')
plt.title("Erosión")
plt.axis('off')
# Adds a subplot
fig.add_subplot(rows, columns, 6)
plt.imshow(dil, cmap='gray')
plt.title("Dilatación")
plt.axis('off')
```

```python
# Imagen final con cuadraditos
# create figure----------------------------------------
fig = plt.figure(figsize=(20, 14))

# Adds a subplot
fig.add_subplot(rows, columns, 1)
plt.imshow(array1, cmap='gray')
plt.axis('off')
plt.title("ORIGINAL")

# apply threshold-------------------------------------
thresh = threshold_otsu(dil.copy())
bw = closing(dil > thresh, square(1))

# remove artifacts connected to image border
cleared = clear_border(bw)
# label image regions
label_image = label(cleared)

# to make the background transparent, pass the value of `bg_label`,
# and leave `bg_color` as `None` and `kind` as `overlay`
image_label_overlay = label2rgb(label_image, image=dil, bg_label=0)
fig, ax = plt.subplots(figsize=(9, 5))
ax.imshow(image_label_overlay)

for region in regionprops(label_image):
  # take regions with large enough areas
  if region.area >= 0.1:
  # draw rectangle around segmented coins
    minr, minc, maxr, maxc = region.bbox
    rect = mpatches.Rectangle((minc, minr), maxc - minc, maxr - minr,fill=False, edgecolor='red', linewidth=2)
    ax.add_patch(rect)

ax.set_axis_off()
plt.tight_layout()
plt.show()
```
